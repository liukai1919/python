```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,1000)
fig = plt.figure()
ax=plt.axes()
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
color=['red','blue','white','black','yellow']
style=['-','-','-.',':',':']
for n in range(5):
    plt.plot(x,np.sin(x-n),color=color[n],linestyle=style[n])
plt.xlim(1,5)
plt.ylim(-0.5,0.5)
```

```python
plt.plot(x,np.sin(x))
plt.title('A,Sine Curve')
plt.xlabel('X')
plt.ylabel('Sin(x)')
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.legend()
```
