```python
def draw_bs_reps(data,func,size=1):
    re_samples=np.empty(size)
    for i in range(size):
        re_samples[i]=func(np.random.choice(data,size=len(data)))
    return re_samples
```

```python
new_sample=draw_bs_reps(shift_b,np.mean,1000000)
p=np.sum(new_sample!=np.mean(force_b))/1000000
print(p)
```
